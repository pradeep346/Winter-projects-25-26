```python
import numpy as np
```

defining the class LinearSVM with the 3 functions defined in question

```python
class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_para=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_para = lambda_para
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        y_ = np.where(y <= 0, -1, 1)
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                margin = y_[idx] * (np.dot(x_i, self.w) - self.b)

                if margin >= 1:
                    dw = 2 * self.lambda_para * self.w
                    db = 0
                else:
                    dw = 2 * self.lambda_para * self.w - y_[idx] * x_i
                    db = y_[idx]
                self.w -= self.lr * dw
                self.b -= self.lr * db

    def predict(self, X):
        linear_output = np.dot(X, self.w) - self.b
        return np.sign(linear_output)
```

constructing a dataset for our problem

```python
X = np.array([
    [2, 3],
    [1, 1],
    [2, 1],
    [3, 2]
])

y = np.array([1, 0, 0, 1])
```

just iterating our given code


```python
svm = LinearSVM(
    learning_rate=0.001,
    lambda_para=0.01,
    n_iters=2000
)

svm.fit(X, y)
```

```python
predictions = svm.predict(X)
print("Predictions:", predictions)
```
