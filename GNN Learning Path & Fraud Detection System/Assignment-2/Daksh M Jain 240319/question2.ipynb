```python
import numpy as np
```

defining activation functions

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
```

the dataset used for XOR operations

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [1], [1], [0]])
```

initializing our parameters


```python
np.random.seed(1)

W1 = np.random.randn(2, 2)
b1 = np.zeros((1, 2))

W2 = np.random.randn(2, 1)
b2 = np.zeros((1, 1))

learning_rate = 0.1
epochs = 8000
```

iterating our required code and updating the parameters

```python
for _ in range(epochs):
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_hat = sigmoid(z2)


    error = y_hat - y
    dW2 = np.dot(a1.T, error * sigmoid_derivative(z2))
    db2 = np.sum(error * sigmoid_derivative(z2), axis=0, keepdims=True)

    d_hidden = np.dot(error * sigmoid_derivative(z2), W2.T)
    dW1 = np.dot(X.T, d_hidden * sigmoid_derivative(z1))
    db1 = np.sum(d_hidden * sigmoid_derivative(z1), axis=0, keepdims=True)

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
```

```python
print("Predicted outputs:")
print(np.round(y_hat))
```
